```python
if 'Total' in df.columns and 'Traffic Situation' in df.columns:
    df['Total'] = pd.to_numeric(df['Total'], errors='coerce')

    df_clean = df.dropna(subset=['Total', 'Traffic Situation'])

    total_by_traffic_situation = df_clean.groupby('Traffic Situation')['Total'].sum()

    plt.figure(figsize=(10, 6))
    total_by_traffic_situation.plot(kind='bar', color='skyblue')
    plt.title('Total Vehicle Count by Traffic Situation')
    plt.xlabel('Traffic Situation')
    plt.ylabel('Total Vehicle Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
if 'Time' in df.columns and 'Total' in df.columns:
    df['Time'] = pd.to_datetime(df['Time'], errors='coerce')

    df['Hour'] = df['Time'].dt.hour

    df['Total'] = pd.to_numeric(df['Total'], errors='coerce')

    df_clean = df.dropna(subset=['Total', 'Hour'])

    total_by_hour = df_clean.groupby('Hour')['Total'].sum()

    plt.figure(figsize=(10, 6))
    total_by_hour.plot(kind='bar', color='skyblue')
    plt.title('Total Vehicle Count by Hour of the Day')
    plt.xlabel('Hour of the Day')
    plt.ylabel('Total Vehicle Count')
    plt.xticks(rotation=0)
    plt.show()
```

```python
if 'Time' in df.columns and 'Total' in df.columns:
    df['Time'] = pd.to_datetime(df['Time'], errors='coerce')

    df['Hour'] = df['Time'].dt.hour

    df['Total'] = pd.to_numeric(df['Total'], errors='coerce')

    df_clean = df.dropna(subset=['Total', 'Hour'])

    total_by_hour = df_clean.groupby('Hour')['Total'].sum().reset_index()

    plt.figure(figsize=(10, 6))
    heatmap_data = total_by_hour.pivot_table(values='Total', index='Hour', aggfunc='sum')
    sns.heatmap(heatmap_data, cmap='Blues', annot=True)
    plt.title('Heatmap of Total Vehicle Count by Hour')
    plt.xlabel('Hour of the Day')
    plt.ylabel('Total Vehicle Count')
    plt.show()
```
